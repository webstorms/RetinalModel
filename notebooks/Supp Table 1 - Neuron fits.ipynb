# Prediction accuracy across different scalings

This notebook generates the table containing the fit accuracy of the encoding model across the different datasets using different scalings of the input dimensions (Supplementary Table 1). All remaining models (trained for prediction) were fit on each dataset using the scale for which the encoding model performed the best.

```python
import os

from retina import analysis

%load_ext autoreload
%autoreload 2
```

```python
root = os.path.expanduser("~/PycharmProjects/RetinalModel")
```

```python
prediction_results = analysis.PredictionResults(root)
```

```python
def to_table_latex(dataset_name, dataset_df):
    dataset_df = dataset_df.groupby(["pred_ms", "dim"]).mean()
    max_cc = "{:.2f}".format(dataset_df.loc[dataset_df["cc_norm"].idxmax()]["cc_norm"])
    s = ""
    
    for i, scale in enumerate([0.66, 1.0, 1.5]):
        cc = "{:.2f}".format(dataset_df.iloc[i]["cc_norm"])
        scale = f"$\\times$ {scale}"
        
        if cc == max_cc:
            scale = "\\bf{" + "{0}".format(scale) + "}"
            cc = "\\bf{" + "{0}".format(cc) + "}"
        
        if i == 0:
            s += f"{dataset_name} & {scale} & {cc}\\\\"
        else:
            s += f"& {scale} & {cc}\\\\"
    
    return s
```

```python
supp_table_str = to_table_latex("Macaque movie", prediction_results.supp_macaque_movie_df) 
supp_table_str += to_table_latex("Macaque image", prediction_results.supp_macaque_image_df)
supp_table_str += to_table_latex("Marmoset movie", prediction_results.supp_marmoset_movie_df)
supp_table_str += to_table_latex("Salamander image", prediction_results.supp_salamander_image_df)
supp_table_str += to_table_latex("Mouse movie", prediction_results.supp_mouse_movie_df)
```
